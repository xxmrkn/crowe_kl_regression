```python
def visualize_scatter(df,df2):
    sns.set(font_scale = 2.5)

    fig, ax = plt.subplots(figsize=(15,10))

    corr = str(df.corr()['True Label']['Probability'])
    corr = corr[:5]

    #mappable = ax.scatter(x=df['True Label'],y=df['Probability'],cmap='coolwarm', s=200)
    sns.regplot(x=df['True Label'],y=df['Probability'],data=df)
    ax.set_title(f'Correlation : {corr}')
    ax.set_yticks([0,1,2,3,4,5,6])
    ax.set_xlabel('Ground Truth')
    ax.set_ylabel('Predicted')

```

```python
visualize_scatter(df,df2)
```
